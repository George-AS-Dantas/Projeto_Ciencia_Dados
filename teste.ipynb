```python
import numpy as np

# 1. Entrada do Usuário
salario = 3000

# 2. Definição da Estratégia (O que você falou: 50%, 20%, 15%, 15%)
# Criamos um array com as fatias do bolo
fatias = np.array([0.50, 0.20, 0.15, 0.15])
nomes_fatias = ['Fixos', 'Investimentos', 'Lazer', 'Variáveis']

# 3. O Cálculo Mágico (Broadcasting)
# Aqui o NumPy multiplica o salário por TODOS os percentuais ao mesmo tempo
limites_em_reais = salario * fatias

print(f"Salário: R$ {salario}")
print("-" * 20)
# Mostrando os resultados
for i in range(4):
    print(f"Meta para {nomes_fatias[i]}: R$ {limites_em_reais[i]:.2f}")
```
